```python
import pandas as pd
```

```python
# import the training data containing 839 tweets
df = pd.read_csv("data/HCR/dev/orig/hcr-dev.csv")
```

```python
df.head()
```

```python
df = df.drop(['user id', 'username', 'target', 'annotator id', 'comment', 'dispute'], axis=1)
```

```python
df = df.reset_index(drop=True)
df = df.dropna()
```

```python
df.head()
```

```python
set(df['sentiment'])
```

```python
for i, row in df.iterrows():
    new = row[1].strip()
    row[1] = new

df = df[df.sentiment != 'irrelevant']
df = df[df.sentiment != 'unsure']
df = df[df.sentiment != ' irrelevant']
df = df[df.sentiment != ' negative']
df = df[df.sentiment != ' neutral']
set(df['sentiment'])
len(df)
```

```python
df
```

```python
# export reformatted data to use for classyfier
df.to_csv("data/HCR/dev/orig/HCR_formatted_dev.csv")
```
