```python
import pandas as pd
```

```python
# df = pd.DataFrame.from_csv("data/OMD/debate.tsv")
with open('data/OMD/debate.tsv') as file:
    doc = file.read()


# Rating Codes: 1 = negative, 2 = positive, 3 = mixed, 4 = other'    
# meeste van een soort telt

lines = doc.split("\n")
start_of_doc = False
only_doc = []
for line in lines:
    splitted_line = line.split()
    if start_of_doc == False and splitted_line[0] == 'tweet.id':
        start_of_doc = True
    if start_of_doc == True:
        only_doc.append(line)
    
```

```python
tweets = []
for line in only_doc:
    split_line = line.split("\t")
    ones = 0
    twos = 0
    threes = 0
    fours = 0 
    if len(split_line[0]) == 9 and split_line[0].isdigit():
        tweet = split_line[2]
        rating_one = line.strip().split()[-1:][0]
        rating_two = line.strip().split()[-2:][0]
        rating_three = line.strip().split()[-3:][0]
        rating_four = line.strip().split()[-4:][0]
        if rating_one.isdigit() == True:
            if int(rating_one) == 1:
                ones += 1
            elif int(rating_one) == 2:
                twos += 1
            elif int(rating_one) == 3:
                threes += 1
            elif int(rating_one) == 4:
                fours += 1
        if rating_two.isdigit() == True:
            if int(rating_two) == 1:
                ones += 1
            elif int(rating_two) == 2:
                twos += 1
            elif int(rating_two) == 3:
                threes += 1
            elif int(rating_two) == 4:
                fours += 1
        if rating_three.isdigit() == True:
            if int(rating_three) == 1:
                ones += 1
            elif int(rating_three) == 2:
                twos += 1
            elif int(rating_three) == 3:
                threes += 1
            elif int(rating_three) == 4:
                fours += 1
        if rating_four.isdigit() == True:
            if int(rating_four) == 1:
                ones += 1
            elif int(rating_four) == 2:
                twos += 1
            elif int(rating_four) == 3:
                threes += 1
            elif int(rating_four) == 4:
                fours += 1
        ratings =  list([ones,twos,threes,fours])
        most_freq_rating = ratings.index(max(ratings))
        if most_freq_rating == 0:
            sentiment = "negative"
        elif most_freq_rating == 1:
            sentiment = "positive"
        elif most_freq_rating == 2:
            sentiment = "neutral"
        elif most_freq_rating == 3:
            sentiment = "neutral"
        #print(ratings, ratings.index(max(ratings)))
        tweets.append(tuple((tweet, sentiment)))
        
        
        
print(len(tweets))
```

```python
neg = 0
pos = 0
neut =0
for tweet in tweets:
    twit, sentiment = tweet
    if sentiment == "negative":
        neg += 1
    elif sentiment == "positive":
        pos += 1
    elif sentiment == "neutral":
        neut += 1
print(pos, neg, neut)
        
```

```python
df_fin = pd.DataFrame(tweets, columns=['tweet', 'sentiment'])
df_fin
```
