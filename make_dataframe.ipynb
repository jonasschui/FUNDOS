```python
json_list = open('geotagged_tweets_20160812-0912.jsons', "r")
counter = 0
df_def = pd.DataFrame()
tweet_info = dict()
lines = []

for line in json_list:
    try:
        obj = json.loads(line)
        text = obj['text']
        tweet_id = obj['id']
        user_name = obj['user']['screen_name']

        followers = obj['user']['followers_count']
        created_at = obj['created_at']
        place = obj['place']['full_name']
        country = obj['place']['country']
        coordinates = obj['place']['bounding_box']['coordinates']
        for i in coordinates:
            for coordinate in i:
                xcoordinate = coordinate[0]
                ycoordinate = coordinate[1]

        user_mentions = obj['entities']['user_mentions']
        user_mention_list = []
        for i in range(len(user_mentions)):
            usr = obj['entities']['user_mentions'][i]['screen_name']
            user_mention_list.append(usr)

        replied_to = obj['in_reply_to_screen_name']
        media_type = obj['extended_entities']['media'][0]['type']

        #with open("check_tweet_data{}.json".format(counter), "w") as write_file:
        #    json.dump(obj, write_file)
        #df = pd.io.json.json_normalize(obj)
        #df_def = df_def.append(df)
        counter += 1
        tweet_info[tweet_id] = [user_name, text, followers, created_at, place, country,
                                    xcoordinate, ycoordinate, user_mention_list, replied_to,media_type]
    except:
        continue
    if counter%1000 == 0:
        print(counter)



    
```

```python
len(tweet_info)
```

```python
df = pd.DataFrame.from_dict(tweet_info, orient='index', columns=['user_name', 'text', 'followers',
                                                                 'created_at', 'place', 'country',
                                    'xcoordinate', 'ycoordinate', 'user_mention_list', 'replied_to','media_type'])
```
